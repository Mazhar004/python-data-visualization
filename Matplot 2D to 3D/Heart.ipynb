# 1. Import Library

```python
import numpy as np
import matplotlib.pyplot as plt
```

# 2. Heart Shape

```python
def sin_val(x): return np.sin(np.deg2rad(x))


def cos_val(x): return np.cos(np.deg2rad(x))


x = np.linspace(0, 360, 360)

m = 16*(sin_val(x)**3)
n = 13*cos_val(x) - 5*cos_val(2*x) - 2*cos_val(3*x) - cos_val(4*x)

color = ['indigo', 'red']
length = len(color)

plt.figure(figsize=(10, 10))
plt.title('Heart')

for i in range(1, 101):
    plt.plot(0.1*i*m, 0.1*i*n, c=color[i % length])

plt.show()
```
