# 1. Import Library

```python
import matplotlib.pyplot as plt
import numpy as np
```

# 2. Random Sine Shape

```python
def sin_val(x): return np.sin(np.deg2rad(x))


def cos_val(x): return np.cos(np.deg2rad(x))


x = np.linspace(0, 360, 360)

color = ['red', 'blue', 'green']
length = len(color)

plt.figure(figsize=(10, 6))
plt.title('Random Sine')

for i in range(1, 101):
    val = np.random.random()
    plt.plot(i*x, val*i*sin_val(x), c=color[i % length])

plt.show()
```
