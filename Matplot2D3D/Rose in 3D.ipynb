```python
import numpy as np
import plotly.graph_objects as go
```

```python
def generate_rose_points(
    resolution=800, 
    petal_width_factor=1.995653, 
    petal_height_factor=1.27689, 
    petal_falloff=8, 
    petals_per_cycle=3.6
):
    """Generate X, Y, Z coordinates for the 3D rose shape."""
    
    # Define radial and angular values
    radius_values = np.linspace(0, 1, resolution)
    angle_values = np.linspace(-2, 20 * np.pi, resolution)
    R, THETA = np.meshgrid(radius_values, angle_values)

    # Shape modifier to control petal curvature
    petal_shape = 1 - 0.5 * ((5/4) * (1 - np.mod(petals_per_cycle * THETA, 2 * np.pi) / np.pi)**2 - 1/4)**2

    # Angle distortion to create a smooth petal falloff effect
    angle_falloff = (np.pi / 2) * np.exp(-THETA / (petal_falloff * np.pi))

    # Petal thickness based on radius and height factors
    petal_thickness = petal_width_factor * (R**2) * (petal_height_factor * R - 1)**2 * np.sin(angle_falloff)

    # Adjust petal shape with curvature
    adjusted_radius = petal_shape * (R * np.sin(angle_falloff) + petal_thickness * np.cos(angle_falloff))
    
    # Convert to Cartesian coordinates
    X = adjusted_radius * np.sin(THETA)
    Y = adjusted_radius * np.cos(THETA)
    Z = petal_shape * (R * np.cos(angle_falloff) - petal_thickness * np.sin(angle_falloff))
    
    return X, Y, Z

def plot_rose(X, Y, Z):
    """Plot the 3D rose using Plotly."""
    
    # Define red color gradient
    color_gradient = [[0, 'rgb(255,0,0)'], [1, 'rgb(64,0,0)']]  # Bright red to dark red
    
    # Create the 3D surface plot
    fig = go.Figure(data=[go.Surface(x=X, y=Y, z=Z, colorscale=color_gradient, showscale=False)])

    # Set camera angle and labels
    fig.update_layout(
        title="3D Rose 🌹",
        scene=dict(
            xaxis_title="Width",
            yaxis_title="Depth",
            zaxis_title="Height",
            camera=dict(eye=dict(x=-1.5, y=-1.5, z=1.5))
        )
    )
    
    fig.show()
```

```python
# Generate and plot the 3D rose
X, Y, Z = generate_rose_points()
```

```python
plot_rose(X, Y, Z)
```
