# 1. Import Library

```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

# 2. Data Generate to Plot

### Prime number generate
### Cartesian to Polar conversion

```python
def prime(n):
    data = [1, 2, 3]
    for i in range(4, n+1):
        flag = 1
        for j in range(i-1, 1, -1):
            if i % j == 0:
                flag = 0
                break
        if flag:
            data.append(i)
    return data


def data2polar(data):
    r, theta = [], []
    for i in data:
        r.append(np.sqrt(i**2+i**2))
        theta.append(i*np.arctan2(i, i))
    return r, theta
```

# 3. Pattern Plot

### Pattern for all same number pair( (1,1),(2,2)...(n,n) ) [ Left Image ]
### Pattern for only prime number pair( (2,2),(3,3),(5,5)(7,7)....) [ Right image ]

```python
def prime_plot(number_ex=300, prime_ex=5000):
    os.makedirs('Images', exist_ok=True)

    r1, theta1 = data2polar(range(1, number_ex+1))
    r2, theta2 = data2polar(prime(prime_ex))

    fig = plt.figure(figsize=(15, 6))

    ax1 = fig.add_subplot(121, projection='polar')
    ax1.scatter(r1, theta1)
    ax1.set_title('All Number Pattern',fontsize=25)

    ax2 = fig.add_subplot(122, projection='polar')
    ax2.scatter(r2, theta2)
    ax2.set_title('Prime Number Pattern',fontsize=25)

    plt.savefig('Images/Prime Number Pattern.png')
    plt.show()
```

```python
prime_plot()
```
