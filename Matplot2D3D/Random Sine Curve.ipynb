# 1. Import Library

```python
import matplotlib.pyplot as plt
import os
import numpy as np
```

# 2. Random Sine Shape

```python
def sin_val(x): return np.sin(np.deg2rad(x))


def cos_val(x): return np.cos(np.deg2rad(x))

def sine_plot(line=100):
    os.makedirs('Images', exist_ok=True)
    
    x = np.linspace(0, 360, 360)

    color = ['red', 'blue', 'green']
    length = len(color)

    plt.figure(figsize=(10, 6))
    plt.title('Random Sine',fontsize=25)

    for i in range(1, line+1):
        val = np.random.random()
        plt.plot(i*x, val*i*sin_val(x), c=color[i % length])
    plt.savefig('Images/Random Sine.png')
    plt.show()
```

```python
sine_plot(100)
```
