# 1. Import Library

```python
import numpy as np
import matplotlib.pyplot as plt
```

# 2. Data Generate to Plot

### Prime number generate
### Cartesian to Polar conversion

```python
def prime(n):
    data = [1, 2, 3]
    for i in range(4, n+1):
        flag = 1
        for j in range(i-1, 1, -1):
            if i % j == 0:
                flag = 0
                break
        if flag:
            data.append(i)
    return data


def data2polar(data):
    r, theta = [], []
    for i in data:
        r.append(np.sqrt(2*i**2))
        theta.append(i*np.deg2rad(1))
    return r, theta


r1, theta1 = data2polar(range(1, 300))
r2, theta2 = data2polar(prime(5000))
```

# 3. Pattern Plot

### Pattern for all same number pair( (1,1),(2,2)...(n,n) ) [ Left Image ]
### Pattern for only prime number pair( (2,2),(3,3),(5,5)(7,7)....) [ Right image ]

```python
fig = plt.figure(figsize=(15, 6))

ax1 = fig.add_subplot(121, projection='polar')
ax1.scatter(r1, theta1)
ax1.set_title('All Number Pattern')


ax2 = fig.add_subplot(122, projection='polar')
ax2.scatter(r2, theta2)
ax2.set_title('Prime Number Pattern')

plt.show()
```
