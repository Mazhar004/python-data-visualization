# 1. Import Library

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

# 2. Data Generate

### Normal Distribution Formula:
<ul style="font-size:25px;font-family:bold;">
    <li>P(x) = Normal Distribution Formmula</li><br>
    <li>To simplify the equation here, &sigma; = 1</li><br>
    <li>P(x) = <sup>1</sup>&frasl;<sub>&#8730;2&pi;</sub>&#215;e<sup><sup>-x<sup>2</sup></sup></sup>&frasl;<sub>2</sub></li><br>
    
    
    
</ul>

```python
def area(x): return ((1/np.sqrt(2*np.pi))*np.exp(-x/2))

def data_gen(size=1000):
    x = np.random.randn(1, size)
    y = np.random.randn(1, size)
    xy = x**2+y**2
    z = 10*area(xy)
    return (x,y,z)

x,y,z = data_gen(2000)
```

# 3. 3D Representation of Normal Distribution

```python
fig = plt.figure(figsize=(20, 10))

ax = fig.add_subplot(1, 1, 1, projection='3d')

ax.scatter(x,y,z)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.title('Normal Distribution')
plt.show()
```
